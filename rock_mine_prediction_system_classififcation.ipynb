```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

DATA COLLECTION AND DATA PROCESSING

```python
# loading the dataset using pandas here
sonar_data=pd.read_csv('/content/Copy of sonar data.csv', header=None)
```

```python
sonar_data.head(5)
```

```python
# know the number of rows and columns

sonar_data.shape
```

```python
sonar_data.describe()  # statistical measures of the data
```

```python
sonar_data.info() # there are no null values in this dataset so good to proceed for classification
```

```python
sonar_data.value_counts()
```

```python
# know the values in the 60th column

sonar_data[60].value_counts()
```

  M = MINE
  R = ROCK

```python
sonar_data.groupby(60).mean()   # mean is very important for us because by this only we can make predict whether it is rock or mine
```

```python
# seperating the data and label
X = sonar_data.drop(columns= 60 , axis=1)
Y = sonar_data[60] # dropping the 60th column in original dataset and storing the 60th column in Y. so that the X if excluded the 60th column and shifted to Y

```

```python
print(X)
print(Y)
```

TRAINING AND TEST DATA

```python
# we gonna use train_test_data module from sklearn.model_selection importing train_test_data
X_train,X_test,Y_train, Y_test = train_test_split(X,Y, test_size= 0.1, stratify=Y, random_state=1)
```

```python
 print(X.shape, X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

MODEL TRAINING >>>> LOGIISTIC REGRESSION


```python
model = LogisticRegression()
# we created the model now we have to train the model with the training data
```

```python
print(X_train)
print(Y_train)
```

```python
#  training the data into the logistic model
model.fit(X_train,Y_train)
```

MODEL EVALUATION


```python
#  checking the accuracy on trained data
# this model now got trained and seen what the data is. now it is yet to see the untrained data  to give the results
# more than 70% accuracy is good

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on the training data :',training_data_accuracy)
```

```python
#  accuracy on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)

```

```python
 print('Accuracy score on the test data :', test_data_accuracy)
```
